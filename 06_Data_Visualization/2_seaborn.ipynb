we use distribution plots to visualize the distribution of quantative data.

Let's deccuss some plot that allow us to vizualize the distribution of a data set. these plot are:

> distplot/histplot
> join plot
> pairplot
> rugplot


# distribution plots:

```python
import seaborn as sns
import matplotlib as plt
```

```python
df = sns.load_dataset("tips")
df
```

```python
df['size'].unique()
```

```python
# plt.subplot(1,2,1) why this is not wrking?
sns.histplot(df['total_bill'], bins=20, kde=True) # kernal density estimation
sns.histplot(df['tip'], bins=20, kde=True)
# plt.subplot(1,2,1) why this is not wrking?
```

```python
# join plot
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg') # regression plot
# use case of joinplot
```

```python
sns.boxplot(x='total_bill', y='tip', data=df)
```

# pair plot

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='size')
```

```python
# rug plot
sns.rugplot(df)
```
