# Implot()

regression plot is used in machine learning so we will not fully discuss what, hwo and where ti is used but we will see it is created and what are the values we can put inside it

```python
import seaborn as sns
import matplotlib as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue='sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue='sex', palette='rainbow')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue='sex', palette='rainbow', markers=['o', 'v'])
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue='sex', palette='rainbow', markers=['o', 'v'], scatter_kws={'s':100})
```
