```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
x

y = x ** 2
y
```

```python
plt.plot(x, y)
plt.title("Square Function")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

# Subplot in matplotlib

```python
plt.subplot(2, 2, 1) # 2 rows, 2 columns, plot 1
plt.plot(x, y, 'r') # plot 1 data   r means red color
plt.subplot(2, 2, 4) # 2 rows, 2 columns, plot 2
plt.plot(x, y) # plot 4 data 
plt.subplot(2, 2, 3) # 2 rows, 2 columns, plot 3
plt.plot(y, x) # plot 3 data
plt.subplot(2, 2, 2) # 2 rows, 2 columns plot 4
plt.plot(x**2, x**2) # plot 2 data
```

```python
# creating 6 by 6 inches figure
plt.subplot(6, 7, 1)
plt.subplot(6, 7, 3)
plt.subplot(6, 7, 5)
plt.subplot(6, 7, 2)
plt.subplot(6, 7, 4)
plt.subplot(6, 7, 8)
plt.subplot(6, 7, 7)
plt.subplot(6, 7, 9)

plt.tight_layout() # to avoid overlap

```

# object oriented way

```python
# usecase of object oriented way: to have more control over the plots, like adding multiple plots in a single figure
# or move the plots around
fig = plt.figure() # creating a figure
axis1 = fig.add_axes([0.1, 0.1, 0.6, 0.8]) # left, bottom, width, height (range 0 to 1)
axis1.plot(x, y, 'b') # plotting on the axis

axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # left, bottom, width, height (range 0 to 1)
axis2.plot(y, x, 'b') # plotting on the axis
```

```python
# dpi and figsize
fig = plt.figure(figsize=(9,4), dpi=100) # figsize in inches, dpi is dots per inch  # higher the dpi, better the quality
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, 'g')
ax.set_title("Title")
```

# Types of plot

```python
# histogram
data = np.random.randn(1000)
plt.hist(data) # histogram
plt.hist(data, bins=30) # histogram with 30 bins
```

```python
# boxtplot
plt.boxplot(x)
```

# saving plots

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
fig.savefig("myfigure.png") # saving the figure

# another way
plt.savefig("myfigure2.png") # saving the figure
```
