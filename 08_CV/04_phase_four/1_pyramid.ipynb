```python
img = cv2.imread('img_5.jpg')
print(img.shape )  # prints (height, width, channels)

new = cv2.pyrDown(img)   # pydown used to reduce the size of image, while pyrup used to increase the size of image
print(new.shape)         # prints (height/2, width/2, channels)

new2 = cv2.pyrDown(new)
print(new2.shape)        # prints (height/4, width/4, channels)

new3 = cv2.pyrDown(new2)
print(new3.shape)        # prints (height/8, width/8, channels)

cv2.imshow('Original Image', img)
cv2.imshow('Pyramid Down', new)
cv2.imshow('Pyramid Down 2', new2)
cv2.imshow('Pyramid Down 3', new3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# now we will increase the size of image using pyrup
up = cv2.pyrUp(new3)
print(up.shape)          # prints (height/4, width/4, channels)

up2 = cv2.pyrUp(up)
print(up2.shape)         # prints (height/2, width/2, channels)

up3 = cv2.pyrUp(up2)
print(up3.shape)        # prints (height, width, channels)

cv2.imshow('Pyramid Up', up)
cv2.imshow('Pyramid Up 2', up2)
cv2.imshow('Pyramid Up 3', up3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
