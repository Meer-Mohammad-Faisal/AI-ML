```python
import cv2
import numpy as np
```

```python
img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (400, 600))


mask = np.float32([[1, 0, 100], 
                   [0, 1, 50]])

new = cv2.warpAffine(img, mask, (400, 600))



cv2.imshow('Original Image', img)
#cv2.imshow('Mask', mask)
cv2.imshow('Translated Image', new)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
