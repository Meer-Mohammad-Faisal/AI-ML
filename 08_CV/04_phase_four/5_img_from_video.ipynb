```python
import cv2
```

# Extracting Image from video

```python
cap = cv2.VideoCapture(r"E:\Kesariya - Brahmastra-(Hd9video).mp4")
c = 0
while True:
    ret, frame = cap.read()
    if ret == True:
        frame = cv2.resize(frame, (800, 600))
        filename = "E://Kesariya - Brahmastra-(Hd9video)org_img_"+str(c) + ".jpg"
        cv2.imwrite(filename, frame)
        
        cv2.imshow("Video Frame", frame)
        c += 1
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
