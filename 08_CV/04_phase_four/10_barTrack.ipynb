track bar


```python
import cv2
import numpy as np
```

```python
def nothing(x):
    pass    
```

```python
img = np.zeros((500, 600, 3), np.uint8) * 255
cv2.namedWindow('bar')
cv2.createTrackbar('on', 'bar', 0, 100, nothing)
while True:
    cv2.imshow('bar', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    on = cv2.getTrackbarPos('on', 'bar')
    img[:] = [0, 0,on]

cv2.destroyAllWindows()
```

```python
cv2.imshow('Image', img)  
cv2.waitKey(0)
cv2.destroyAllWindows()
```
