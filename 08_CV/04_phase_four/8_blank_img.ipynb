```python
import cv2
import numpy as np
```

```python
img = cv2.imread('img_1.jpg')
img = cv2.resize(img, (600, 500))

print(img.shape)
```

```python
print(img)
```

```python
new_img = np.ones((500, 600, 3), dtype=np.uint8) * 255
#print(new_img)

cv2.imshow('Blank Image', new_img)
cv2.imshow('Image', img)  
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_img = np.zeros((500, 600, 3), dtype=np.uint8) * 255
#print(new_img)

cv2.imshow('Blank Image', new_img)
cv2.imshow('Image', img)  
cv2.waitKey(0)
cv2.destroyAllWindows()
```
