```python
import cv2
```

```python
org = cv2.imread('img_5.jpg')
org = cv2.resize(org, (400, 600))

new = cv2.cvtColor(org, cv2.COLOR_BGR2GRAY)
new2 = cv2.cvtColor(org, cv2.COLOR_BGR2RGB)
new3 = cv2.cvtColor(org, cv2.COLOR_BGR2HSV)
new4 = cv2.cvtColor(org, cv2.COLOR_BGR2LAB)
new5 = cv2.cvtColor(org, cv2.COLOR_BGR2LUV)
new6 = cv2.cvtColor(org, cv2.COLOR_BGR2YUV)

cv2.imshow('Original Image', org)
cv2.imshow('Gray Image', new)
cv2.imshow('RGB Image', new2)
cv2.imshow('HSV Image', new3)
cv2.imshow('LAB Image', new4)
cv2.imshow('LUV Image', new5)
cv2.imshow('YUV Image', new6)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```
