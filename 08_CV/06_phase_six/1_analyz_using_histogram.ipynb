```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
blank_w = np.ones((500, 500, 3), np.uint8)*255
blank_b = np.zeros((500, 500, 3), np.uint8)*255
```

```python
# Histogram:
# convert to grayscale so calcHist gets a single-channel image
gray = cv2.cvtColor(blank_w, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

plt.plot(hist)
plt.title('Grayscale histogram')
plt.xlim([0, 256])
plt.show()

# display images inline (safer in Jupyter than cv2.imshow)
plt.figure(figsize=(4,4))
plt.imshow(cv2.cvtColor(blank_w, cv2.COLOR_BGR2RGB))
plt.title('blank w')
plt.axis('off')
plt.show()

# #if you want to show blank_b as well:
# plt.figure(figsize=(4,4))
# plt.imshow(cv2.cvtColor(blank_b, cv2.COLOR_BGR2RGB))
# plt.title('blank b')
# plt.axis('off')
# plt.show()
```

```python
# Histogram:
# convert to grayscale so calcHist gets a single-channel image
gray = cv2.cvtColor(blank_b, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

plt.plot(hist)
plt.title('Grayscale histogram')
plt.xlim([0, 256])
plt.show()




#if you want to show blank_b as well:
plt.figure(figsize=(4,4))
plt.imshow(cv2.cvtColor(blank_b, cv2.COLOR_BGR2RGB))
plt.title('blank b')
plt.axis('off')
plt.show()
```

```python


# Read images
img1 = cv2.imread("img_5 copy.jpg")
img2 = cv2.imread("output.jpg")
img3 = cv2.imread("thres.png")

# Convert to grayscale for histogram
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

# Histograms
hist1 = cv2.calcHist([gray1], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([gray2], [0], None, [256], [0, 256])
hist3 = cv2.calcHist([gray3], [0], None, [256], [0, 256])

# Plotting histograms
plt.figure(figsize=(8, 10))

plt.subplot(3, 1, 1)
plt.title("Histogram 1")
plt.plot(hist1)

plt.subplot(3, 1, 2)
plt.title("Histogram 2")
plt.plot(hist2)

plt.subplot(3, 1, 3)
plt.title("Histogram 3")
plt.plot(hist3)

plt.tight_layout()
plt.show()

# OpenCV windows
cv2.imshow("1st", img1)
cv2.imshow("2nd", img2)
cv2.imshow("3rd", img3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
