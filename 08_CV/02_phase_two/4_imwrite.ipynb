```python
import cv2
import numpy as np
```

```python
img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (200, 300))
re_img = cv2.resize(img, (200, 300))

h = np.hstack((re_img, img))
v = np.vstack((h, h))

cv2.imwrite('new_immg.jpg', v)

cv2.imshow('Resized and Original Image', v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
