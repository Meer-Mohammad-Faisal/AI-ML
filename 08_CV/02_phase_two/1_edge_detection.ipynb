```python
import cv2
```

```python
img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (400, 600))
print(img.shape)

new_img = cv2.Canny(img, 100, 200, apertureSize=5, L2gradient=True)
print(new_img.shape)

cv2.imshow('Original Image', img)
cv2.imshow('Edge Detected Image', new_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
