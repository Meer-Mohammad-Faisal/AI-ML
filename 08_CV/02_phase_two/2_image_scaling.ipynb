Rotation

```python
org_img = cv2.imread('img_5.jpg')
res_img = cv2.resize(org_img, (400, 600)) 
w,h = res_img.shape[0], res_img.shape[1]

m = cv2.getRotationMatrix2D((h/2,w/2),80,1)
new_img = cv2.warpAffine(res_img,m,(h,w))


h = np.hstack((res_img, new_img))
cv2.imshow('Rotated Image', h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
