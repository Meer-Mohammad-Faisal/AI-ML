```python
import cv2
import numpy as np
```

```python
img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (400, 600))

img1 = cv2.copyMakeBorder(img,10,10,10,10,cv2.BORDER_CONSTANT,value=[255,0,0])
img1 = cv2.resize(img1, (400, 600))

h = np.hstack((img, img1))

cv2.imshow('Image with Border', h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
