```python
import cv2
```

```python
img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (400, 600))

# fliping y axis
img = cv2.flip(img, 1)

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (400, 600))

# fliping x axis
img = cv2.flip(img, 0)

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (400, 600))


img = cv2.flip(img, -1) # fliping both axis

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Rotation

```python
img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (400, 600))

img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Transition

```python
img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (400, 600))

img = cv2.transpose(img)

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
