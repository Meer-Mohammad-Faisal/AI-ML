```python
import cv2
import numpy as np

img = cv2.imread('img_5.jpg')
img = cv2.resize(img, (300, 450))

def nothing(x):
    pass

# Create window FIRST
cv2.namedWindow("demo")

# Trackbars
cv2.createTrackbar("lb", "demo", 0, 255, nothing)
cv2.createTrackbar("lg", "demo", 0, 255, nothing)
cv2.createTrackbar("lr", "demo", 0, 255, nothing)

cv2.createTrackbar("ub", "demo", 255, 255, nothing)
cv2.createTrackbar("ug", "demo", 255, 255, nothing)
cv2.createTrackbar("ur", "demo", 255, 255, nothing)

while True:
    hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    Lb = cv2.getTrackbarPos("lb", "demo")
    Lg = cv2.getTrackbarPos("lg", "demo")
    Lr = cv2.getTrackbarPos("lr", "demo")

    Ub = cv2.getTrackbarPos("ub", "demo")
    Ug = cv2.getTrackbarPos("ug", "demo")
    Ur = cv2.getTrackbarPos("ur", "demo")

    lo = np.array([Lb, Lg, Lr])
    up = np.array([Ub, Ug, Ur])

    mask = cv2.inRange(hsv_img, lo, up)
    res = cv2.bitwise_and(img, img, mask=mask)

    cv2.imshow("res", res)
    cv2.imshow("mask", mask)
    cv2.imshow("hsv", hsv_img)
    cv2.imshow("Image", img)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# DESTROY WINDOWS AFTER LOOP
cv2.destroyAllWindows()
```
