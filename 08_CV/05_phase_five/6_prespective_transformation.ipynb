```python
import cv2
import numpy as np
```

```python
img = cv2.imread("note.jpg")


cv2.circle(img, (32,160), 6, (0,0,255), -1)
cv2.circle(img, (776, 175), 6, (0,0,255), -1)
cv2.circle(img, (886, 1342), 6, (0,0,255), -1)
cv2.circle(img, (131, 1419), 6, (0,0,255), -1)


cv2.imshow("Notes", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#print(img.shape)
img = cv2.imread('note.jpg')

w, h = (300, 900)
src1 = np.float32([[32,160], [776, 175], [86, 1342],[131, 1419]])
dst1 = np.float32([[0,0], [w,0], [0,h], [w,h]])
m = cv2.getPerspectiveTransform(src1, dst1)
new_img = cv2.warpPerspective(img, m, (w, h))


cv2.imshow("ws", new_img)
cv2.imshow("Notes", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
