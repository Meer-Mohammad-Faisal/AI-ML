saving image

```python
img = cv2.imread('img_person.jpg')
img = cv2.resize(img, (600, 600))



cv2.imwrite('output.jpg', img)
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

croping head of that saved image

```python
# structure of croping: img[y1:y2, x1:x2]
import cv2

img = cv2.imread('output.jpg')
h, w, c = img.shape

# Crop
crop = img[40:230, 200:360]    # (190,160,3)

# First paste → always safe
img[40:230, 360:520] = crop

# Second paste → only paste if space exists
if w >= 680:
    img[40:230, 520:680] = crop
else:
    print("Image too small for second paste!")

cv2.imshow("Crop", crop)
cv2.imshow("Modified Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
