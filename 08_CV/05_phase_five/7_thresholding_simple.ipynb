```python
import cv2
```

```python
img = cv2.imread("thres.png")
img = cv2.resize(img, (400, 400))

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, th = cv2.threshold(img, 190, 255, cv2.THRESH_BINARY)


cv2.imshow("Threshold binary", th)
cv2.imshow("IMAGE", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("thres.png")
img = cv2.resize(img, (400, 400))

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, th = cv2.threshold(img, 190, 255, cv2.THRESH_BINARY_INV)
_, th = cv2.threshold(img, 190, 255, cv2.THRESH_MASK)
_, th = cv2.threshold(img, 190, 255, cv2.THRESH_TRIANGLE)
_, th = cv2.threshold(img, 190, 255, cv2.THRESH_TOZERO)
_, th = cv2.threshold(img, 190, 255, cv2.THRESH_BINARY_INV)


cv2.imshow("Threshold binary", th)
cv2.imshow("IMAGE", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
