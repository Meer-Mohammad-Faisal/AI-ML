```python
import cv2
```

```python
old_img = cv2.imread('img_2.jpg')
old_img = cv2.resize(old_img, (600, 700))

new_img = cv2.line(old_img, (150, 140), (430, 190), (0, 255, 0), 5, 4)

cv2.imshow('Old Image', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

rectangle

```python
old_img = cv2.imread('img_2.jpg')
old_img = cv2.resize(old_img, (600, 700))

txt_img = cv2.putText(old_img, 'Vijay', (90, 30),2, 1, (0, 255, 255), 2, 16, bottomLeftOrigin=False )

new_img = cv2.rectangle(txt_img, (90, 40), (490, 450), (0, 255, 0), 5, 4)

cv2.imshow('Old Image', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
