```python
import cv2
```

```python
old_img = cv2.imread('img_5.jpg')
old_img = cv2.resize(old_img, (600, 900))

new_img = cv2.circle(img = old_img, center=(280, 270), radius=140, color=(255, 0, 0), thickness=5, lineType=8)

cv2.imshow('old_img', old_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
old_img = cv2.imread('img_6.jpg')
old_img = cv2.resize(old_img, (600, 1000))

txt_img = cv2.putText(old_img, 'Kriti', (360, 137),2, 1, (0, 255, 0), 2, 16, bottomLeftOrigin=False )

new_img = cv2.ellipse(img = old_img, center=(320, 125), axes = (50, 90), angle=20, startAngle=0, endAngle=340, color=(0, 255, 0), thickness=3, lineType=16)

cv2.imshow('old_img', old_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
