```python
import cv2
import numpy as np
```

```python
img1 = cv2.imread('binary_img1.png')
img2 = cv2.imread('binary_img2.png')

img1 = cv2.resize(img1, (500, 500))
img2 = cv2.resize(img2, (500, 500))

new = cv2.bitwise_and(img1, img2)


h = np.hstack((img1, img2, new))
cv2.imshow('Bitwise AND Operation', h)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread('binary_img1.png')
img2 = cv2.imread('binary_img2.png')

img1 = cv2.resize(img1, (300, 300))
img2 = cv2.resize(img2, (300, 300))

new = cv2.bitwise_not(img1)
new2 = cv2.bitwise_not(img2)

h = np.hstack((img1, new))
h2 = np.hstack((img2, new2))

v = np.vstack((h, h2))

cv2.imshow('Bitwise NOT Operation', v)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
