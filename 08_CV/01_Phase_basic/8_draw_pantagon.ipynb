```python
import cv2
import numpy as np

# Read and resize image
old_img = cv2.imread('img_5.jpg')
old_img = cv2.resize(old_img, (600, 900))

# Polygon points
pts = np.array([[180, 300],
                [170, 200],
                [360, 300],
                [300, 500],
                [200, 500]], np.int32)

# # Reshape for polylines â†’ (n_points, 1, 2)
# pts = pts.reshape((-1, 1, 2))

# Draw polygon
cv2.polylines(old_img, [pts], isClosed=True, color=(0, 255, 0), thickness=4, lineType=cv2.LINE_AA)

# Show image
cv2.imshow("Polygon Image", old_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
