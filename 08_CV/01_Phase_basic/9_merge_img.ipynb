```python
import cv2
```

```python
img1 = cv2.imread('img_3.jpg')
img2 = cv2.imread('img_5.jpg')

img1 = cv2.resize(img1, (500, 500))
img2 = cv2.resize(img2, (500, 500))


new_img = cv2.addWeighted(img1, 0.7, img2, 0.3, 1)
# new_img = cv2.subtract(img1, img2)
 

cv2.imshow('Blended Image', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
