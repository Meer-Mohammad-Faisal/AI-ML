```python
import cv2
```

```python
img = cv2.imread("img_1.jpg", 0) #BGR (0-255, 0-255, 0-255);  0 for grayscale(black/white), 1 for normal color, -1 for alpha color
# print(img)
print(img.shape)
img = cv2.resize(img, (500,500))
cv2.imshow("Varun", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
