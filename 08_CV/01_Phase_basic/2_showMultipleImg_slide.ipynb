```python
# horisontal stacking image
img = cv2.imread('img_4.jpg') # Read image
re_img = cv2.resize(img, (300, 300)) # Resize image to 500x500

h = np.hstack((re_img, re_img)) # Horizontal stacking

cv2.imshow('Actor Image', h) # Show image in a window
cv2.waitKey(0) # Wait for a key press to close the window. time -> ms
cv2.destroyAllWindows() # Close all OpenCV windows
```

```python
# vertical stacking image
img = cv2.imread('img_4.jpg') # Read image
re_img = cv2.resize(img, (300, 300)) # Resize image to 500x500

v = np.vstack((re_img, re_img, re_img, re_img)) # Vertical stacking

cv2.imshow('Actor Image', v) # Show image in a window
cv2.waitKey(0) # Wait for a key press to close the window. time -> ms
cv2.destroyAllWindows() # Close all OpenCV windows
```

combinin both horizontally and vertically

```python
img = cv2.imread('img_2.jpg') # Read image
re_img = cv2.resize(img, (300, 400)) # Resize image to 200x200
h = np.hstack((re_img, re_img, re_img)) # Horizontal stacking
v = np.vstack((h, h)) # Vertical stacking
cv2.imshow('Actor Image', v) # Show image in a window
cv2.waitKey(0) # Wait for a key press to close the window. time -> ms
cv2.destroyAllWindows() # Close all OpenCV windows
```
