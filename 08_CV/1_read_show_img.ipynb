```python
import cv2
```

Read Imgage

```python
img = cv2.imread('img_2.jpg') # Read image
```

```python
cv2.imshow('Actor Image', img) # Show image in a window
cv2.waitKey(0) # Wait for a key press to close the window. time -> ms
cv2.destroyAllWindows() # Close all OpenCV windows
# cv2.destroyWindow('Actor Image') # Close specific window 
```

```python
cv2.imshow('Actor Image', img) # Show image in a window
cv2.imshow('Actor Image 2', img) # Show image in another window
cv2.waitKey(0) # Wait for a key press to close the window. time -> ms
cv2.destroyAllWindows() # Close all OpenCV windows
#cv2.destroyWindow('Actor Image') # Close specific window 
```
