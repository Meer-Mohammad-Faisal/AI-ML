```python
import cv2
import numpy as np
```

```python
img = cv2.imread('morphological_operations.png')
img = cv2.resize(img, (600, 400))

# Top Hat
m = np.ones((7,7), np.int8)
topHead = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, m, iterations=1)
cv2.imshow('Tophead', topHead)

cv2.imshow('Top head Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('morphological_operations.png')
img = cv2.resize(img, (600, 400))

# Black Hat
m = np.ones((10,10), np.int8)
blackHat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, m, iterations=1)
cv2.imshow('Blackhat image', blackHat)

cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
