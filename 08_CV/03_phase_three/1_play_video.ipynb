```python
import cv2
```

```python
cap = cv2.VideoCapture(r"E:\Kesariya - Brahmastra-(Hd9video).mp4")

while cap.isOpened():
    r,frame = cap.read()
    if r == True:
        frame = cv2.resize(frame, (800, 500))
        cv2.imshow('song', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
