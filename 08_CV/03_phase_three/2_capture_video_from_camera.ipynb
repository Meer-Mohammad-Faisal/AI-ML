```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
while True:
    r, frame = cap.read()
    if r == False:
        break
    else:
        frame = cv2.resize(frame, (640, 480))
        cv2.imshow('Webcam', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
