```python
import cv2
import numpy as np
```

```python
img = cv2.imread('morphological_operations.png')
img = cv2.resize(img, (600, 400))

# gradient
m = np.ones((4,4), np.int8)

gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, m, iterations=1)
cv2.imshow('Gradient image', gradient)

cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
