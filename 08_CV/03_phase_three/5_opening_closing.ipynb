```python
import cv2
import numpy as np
```

```python
img = cv2.imread('morphological_operations.png')
img = cv2.resize(img, (600, 400))

# Opening
m = np.ones((7,7), np.int8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, m, iterations=1)
cv2.imshow('Opening', opening)

cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('morphological_operations.png')
img = cv2.resize(img, (600, 400))

# closing
m = np.ones((7,7), np.int8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, m, iterations=1)
cv2.imshow('Closing', closing)

cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
