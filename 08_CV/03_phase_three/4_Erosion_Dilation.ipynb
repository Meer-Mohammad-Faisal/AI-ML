```python
import cv2
import numpy as np
```

```python
img = cv2.imread('morphological_operations.png')

m = np.ones((7,7), np.int8)

# Erosion
erosion = cv2.erode(img, m, iterations=1)
cv2.imshow('Erosion', erosion)

cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('morphological_operations.png')

m = np.ones((7,7), np.int8)

# Dilation
dilation = cv2.dilate(img, m, iterations=1)
cv2.imshow('Dilation', dilation)

cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
