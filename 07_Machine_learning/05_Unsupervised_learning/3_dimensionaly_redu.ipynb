# PCA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# there will be 5 features and we convert into 2 features using PCA
X, y = make_blobs(n_samples=500, centers=3, n_features=5, cluster_std=1.5, random_state=42)
```

```python
X
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# now we apply PCA-> we want to convert 5 features into 2 features
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
# ploting on 2D graph
df_pca = pd.DataFrame(X_pca, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['label'] = y
```

```python
sns.scatterplot(data=df_pca, x='Principal Component 1', y='Principal Component 2', hue='label', palette='Set2')
```
