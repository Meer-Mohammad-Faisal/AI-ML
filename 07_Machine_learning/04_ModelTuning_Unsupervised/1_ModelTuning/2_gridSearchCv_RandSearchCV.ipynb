```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
# we are going to predict species.
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species', axis = 1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=20)
```

```python
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test, y_test)
# currently it giving me 98%, but when we change n_neighbors from 5 to 20
# then it give 100% socre
```

# Using SVM

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma='auto')
```

```python
model_svm.fit(X_train, y_train)
```

```python
model_svm.score(X_test, y_test)
# currenty it giving 98. but when we change C = 30 to C = 10
# then it give me 100
```

```python
# so, we have to use the best hyperparameter
# that's why we will use grid search cv.
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svm), {
    'C': [1, 10, 20, 30],
    'kernel' : ['fbf', 'linear'],
}, cv = 5, return_train_score=False)
```

```python
classifier.fit(X, y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C', 'param_kernel', 'mean_test_score']]
```

now applying gridSearch cv on KNN

```python
classifier_knn = GridSearchCV((model_knn), {
    'n_neighbors': [10, 5, 20, 32],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'weights': ['uniform', 'distance']

},  cv = 5, return_train_score=False)
```
