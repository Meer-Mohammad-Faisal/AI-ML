```python
import re
```

```python
def  remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

```python
text1 = 'chek https://www.kaggle.com/campusx/notebook34434'
text2 = 'check http://www.kaggle.com/campusx/notebook3'
text3 = 'google search here www.google.com'
text4 = 'for notebook click https://www.kaggle.com/campusx/notebook33ab to search'
```
