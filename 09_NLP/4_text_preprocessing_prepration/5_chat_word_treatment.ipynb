```python
chat = "'asap': As soon As possible"
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat:
            new_text.append(chat[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
chat_conversion('asap do this')
```
