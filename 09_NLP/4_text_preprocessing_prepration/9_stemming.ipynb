```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return ' '.join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walking walked walks"
stem_words(sample)
```

```python
text = "running runner runs easily fair fairly fairness sofia fairless"
stem_words(text)
```
