```python
with open("students.txt") as f:
    text = f.readlines()
text
```

```python
text = ' '.join(text)
text
```

extracting emails of student

```python
doc = nlp(text)
emails = []
for token in doc:
    if token.like_email:
        emails.append(token.text)

emails
```

```python
from spacy.symbols import ORTH
nlp.tokenizer.add_special_case("gimme", [
    {ORTH: "gim"},
    {ORTH: "me"}
])


doc = nlp("gimme dobule cheese extra large healthy pizza")

tokens = [token.text for token in doc]
tokens
```
